```python
from fbprophet import Prophet
```

```python
model = Prophet(interval_width = 0.95)
```

- by default facebook create with 80% confidence interval but here i am giving 95% confidencnce interval.

- by defualt seasonality is auto if you don't give facebook will find itself.but here i know my data is yearly seaonality base.

```python
model.fit(df_train) 
```

```python
model.params
```

i have 40 rows for test our data.so now i am taking model.make_future_dataframe .it will create date object with 40 date.

```python
future = model.make_future_dataframe(periods = 40)
future
```

```python
forecast = model.predict(future)
forecast.head()
```

```python
forecast.tail()
```

```python
forecast[['ds' , 'yhat' , 'yhat_lower' , 'yhat_upper']]
```

Now i am going concat actual close value and predicted close value.

```python
y_pred = pd.concat( [final_df.set_index('ds')['y'] , forecast.set_index('ds')['yhat']] , 1 )
y_pred.head()
```

```python
y_pred.tail(40)
```
